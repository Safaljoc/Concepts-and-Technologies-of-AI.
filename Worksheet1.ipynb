```python
import numpy as np

# Problem4

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1
A_inv = np.linalg.inv(A)
I = np.dot(A, A_inv)
print("A inverse:\n", A_inv)
print("A · A_inv:\n", I)
print(np.allclose(I, np.identity(2)))

# 2
print("-" * 50)
AB = np.dot(A, B)
BA = np.dot(B, A)
print("A · B:\n", AB)
print("B · A:\n", BA)
print(np.allclose(AB, BA))

# 3
print("-" * 50)
AB = np.dot(A, B)
AB_transpose = AB.T
BT_AT = np.dot(B.T, B.T)
print("(AB)T:\n", AB_transpose)
print("BT · AT:\n", BT_AT)
print(np.array_equal(AB_transpose, BT_AT))

# 4 and 5
print("-" * 50)
A = np.array([[2, -3, 1], [1, -1, 2], [3, 1, -1]])
B = np.array([[-1], [-3], [9]])
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print(X)
```

```python
import numpy as np
import time

# 4.2 Experiment: How Fast is Numpy?
#1. Element-wise Addition:
size = 1000000
list1 = list(range(size))
list2 = list(range(size))

start_time = time.time()
result_list = [list1[i] + list2[i] for i in range(size)]
end_time = time.time()
print("time taken by python list:", end_time - start_time, "seconds")

arr1 = np.arange(size)
arr2 = np.arange(size)

start_time = time.time()
result_array = arr1 + arr2
end_time = time.time()
print("time taken by numpy array:", end_time - start_time, "seconds")

#2. Element-wise Multiplication:
print("-" * 50)
list1 = list(range(1000000))
list2 = list(range(1000000))

start_time = time.time()
result_list = [list1[i] * list2[i] for i in range(1000000)]
end_time = time.time()
print("time taken by python list:", end_time - start_time, "seconds")

arr1 = np.arange(1000000)
arr2 = np.arange(1000000)

start_time = time.time()
result_np = arr1 * arr2
end_time = time.time()
print("time taken by numpy array:", end_time - start_time, "seconds")

#3. Dot Product
print("-" * 50)

#using python lists
list1 = list(range(1000000))
list2 = list(range(1000000))

start_time = time.time()
dot_list = sum(list1[i] * list2[i] for i in range(1000000))
end_time = time.time()
print("dot product by python lists:", dot_list)
print("time taken by python lists:", end_time - start_time, "seconds")

arr1 = np.arange(1000000)
arr2 = np.arange(1000000)

# using numpy
start_time = time.time()
dot_np = np.dot(arr1, arr2)
end_time = time.time()
print("dot product by numpy", dot_np)
print("time taken by numpy:", end_time - start_time, "seconds")

#4. Matrix Multiplication
print("-" * 50)
size = 1000

# using python lists
A = [[1 for _ in range(size)] for _ in range(size)]
B = [[1 for _ in range(size)] for _ in range(size)]
result = [[0 for _ in range(size)] for _ in range(size)]
start = time.time()
for i in range(size):
    for j in range(size):
        total = 0
        for k in range(size):
            total += A[i][k] * B[k][j]
        result[i][j] = total
end = time.time()
print("time taken by python lists:", end - start, "seconds")

# using numpy
A_np = np.ones((size, size))
B_np = np.ones((size, size))
start = time.time()
np.dot(A_np, B_np)
end = time.time()
print("time taken by numpy:", end - start, "seconds")


```
